```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
import pandas as pd
```

```python
with open('healthy_old_young.json') as f:
    data = json.load(f)
```

```python
data.keys()
```

```python
def frame_func(data):
    df = pd.DataFrame(data)
    df = df.rename(columns={0: "Healthy", 
                     1: "Incubating",
                     2: "Sick",
                     3: "Recovered",
                     4: "Dead",
                     5: "Time"})
    df.set_index("Time")
    
    return df
```

```python
# Plotjes
healthy = frame_func(data['Healthy'])
old = frame_func(data['Old'])
young = frame_func(data['Young'])
```

```python
plt.stackplot(healthy.index, healthy.Recovered, healthy.Incubating, healthy.Sick, healthy.Dead, colors = ['#13C000', '#B29D00', '#B22E00', '#5B5B5B'])
```

```python
plt.stackplot(old.index, old.Recovered, old.Incubating, old.Sick, old.Dead, colors = ['#13C000', '#B29D00', '#B22E00', '#5B5B5B'])
```

```python
plt.stackplot(young.index, young.Recovered, young.Incubating, young.Sick, young.Dead, colors = ['#13C000', '#B29D00', '#B22E00', '#5B5B5B'])
```
